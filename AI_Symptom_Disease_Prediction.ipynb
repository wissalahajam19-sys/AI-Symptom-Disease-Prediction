```python
import pandas as pd

df = pd.read_csv("Training.csv")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
print(df.shape)
print(df['prognosis'].unique())
```

```python
selected = ['Flu','Cold','Migraine','Allergy']

df = df[df['prognosis'].isin(selected)]
```

```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

```python
def symptoms_to_text(row):
    return ' '.join([col for col in X.columns if row[col]==1])

X_text = X.apply(symptoms_to_text, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)


label_map = dict(zip(le.classes_, le.transform(le.classes_)))
print(label_map)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_text, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


clf = RandomForestClassifier(n_estimators=100, random_state=42)


clf.fit(X_train_vect, y_train)


y_pred = clf.predict(X_test_vect)
```

```python
# Accuracy
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.2f}")

# Classification report
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=le.classes_, yticklabels=le.classes_, cmap="Blues")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

```python
def predict_symptoms(text):
    vect = vectorizer.transform([text])
    pred = clf.predict(vect)
    return le.inverse_transform(pred)[0]

predict_symptoms("itching skin_rash nodal_skin_eruptions")
print("Predicted disease:", le.inverse_transform(pred)[0])
```

```python
import joblib

joblib.dump(clf, "disease_model.pkl")
joblib.dump(vectorizer, "vectorizer.pkl")
joblib.dump(le, "label_encoder.pkl")
```
